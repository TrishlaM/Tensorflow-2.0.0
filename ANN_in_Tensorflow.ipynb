For this project, I used the Fashion MNIST dataset that is sued to categorize garments based on their images. Such classification tasks can be used by e-retailers to automatically classify their garments instead of manually inserting them into each category. To do the classification, a basic shallow artificial neural network had been built using tensorflow and keras.

```
import numpy as np
import tensorflow as tf
import datetime as dt
from tensorflow.keras.datasets import fashion_mnist
```

###**Data Preprocessing**



**1. Load the datasets**


```
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

**2. Normalize he images**


```
X_train = X_train/255.0
X_test = X_test/255.0
```

```
X_train[0][10]
```

**3. Reshape the dataset**

```
X_train = X_train.reshape(-1, 28*28)
X_train.shape
```

```
X_test = X_test.reshape(-1, 28*28)
X_test.shape
```

##**Building the artificial neural network**

**1. Define the model**

```
model = tf.keras.models.Sequential()
```

**2. Add two fully connected layers**
 
 -no. of neurons = 128
  
  -activation function = Relu
  
  -input shape = (784,)

```
model.add(tf.keras.layers.Dense(units = 128, activation = 'relu', input_shape = (784, )))
```

```
model.add(tf.keras.layers.Dense(units = 64, activation = 'relu'))
```

**2. Add another layer with dropout**


```
model.add(tf.keras.layers.Dropout(0.3))
```

**3. Add the output layer**

```
model.add(tf.keras.layers.Dense(units = 10, activation = 'softmax'))
```

**4. Compile the model**

```
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['sparse_categorical_accuracy'])
model.summary()
```

**5. Train the model**

```
model.fit(X_train, y_train, epochs = 500, batch_size = 50000)
```

**5. Evaluate the model**



```
test_loss, test_acc = model.evaluate(X_test, y_test)
```

```
print(test_loss)
print(test_acc)
```
